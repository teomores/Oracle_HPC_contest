```python
def isin_function(pred, linked):
    res = np.empty(len(linked), dtype=int)
    for i in tqdm(range(len(linked))):
        res[i] = np.isin(linked[i], pred[i])
    return res

sub['correct'] = isin_function(sub.predicted_record_id.values, sub.linked_id.values) 
```

```python
from scipy import *
from scipy.sparse import *
import os
```

```python
# Forse Ã¨ meglio usare la cosine similarity?
sim_path = '../dataset/original/similarities'
sim_name = load_npz(os.path.join(sim_path, 'jaccard_uncleaned_name_300k_original_2ngrams.npz'))
```

```python
def n_same_name(name_3nan_idx, sim_name):
    res = np.empty(len(name_3nan_idx), dtype=int)
    for i in tqdm(range(len(name_3nan_idx))):
        res[i] = np.argwhere(sim_name[name_3nan_idx[i]] >= 0.99).shape[0]
    return res
```

```python
df_test_3nan['n_samename_intrain'] = n_same_name(df_test_3nan.index, sim_name)
```

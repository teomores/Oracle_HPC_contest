```python
import pandas as pd
import numpy as np
from tqdm.auto import tqdm
```

```python
df_test = pd.read_csv("../dataset/original/test.csv", escapechar="\\")
sub = pd.read_csv("/Users/alessiorussointroito/Downloads/sub_prova.csv.xls")
```

```python
df_test['linked_id'] = df_test.record_id.str.split("-")
df_test['linked_id'] = df_test.linked_id.apply(lambda x: x[0])
```

```python
sub['predicted_record_id'] = sub.predicted_record_id.str.split(" ")
```

```python
sub = sub.merge(df_test[['record_id', 'linked_id']], how='left', left_on='queried_record_id', right_on="record_id")
```

```python
sub
```

```python
def isin_function(pred, linked):
    res = np.empty(len(linked), dtype=int)
    for i in tqdm(range(len(linked))):
        res[i] = np.isin(linked[i], pred[i])
    return res

sub['correct'] = isin_function(sub.predicted_record_id.values, sub.linked_id.values) 
```

```python
df_train = pd.read_csv("../dataset/original/train.csv", escapechar="\\")
```

```python
df_train['linked_id'] = df_train.linked_id.astype(int)
df_test['linked_id'] = df_test.linked_id.astype(int)
only_test = set(df_test.linked_id.values) - set(df_train.linked_id.values) 
```

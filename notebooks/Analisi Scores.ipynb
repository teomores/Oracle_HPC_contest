```python
restricted_pred = []
max_delta = 2.0
for (q, sc, rec) in tqdm(zip(s.queried_record_id, s.ordered_scores, s.ordered_record)):
    for x in range(len(sc)):
        if x == 0:
            restricted_pred.append((q, sc[x], rec[x]))
        else:
            if x >= 10:
                continue
            elif sc[0] - sc[x] < max_delta:
                restricted_pred.append((q, sc[x], rec[x]))
            else:
                continue
```

```python
restricted_df = pd.DataFrame(restricted_pred, columns=['queried_record_id', 'scores', 'predicted_record_id'])
```
